```python
import pandas as pd
df=pd.read_csv('Covid.csv')#importing file
df.head()#checking first values of dataset
# df.shape #checking number of rows and columns
```

```python
'''Show the number of Confirmed , Deaths and Recovered cases in each 
 region [ use groupby function ]'''

df.groupby(['Confirmed','Region']).count()

```

```python
# 2. Remove All the records where Confirmed cases is less than < 10 [ use ~ ]
df[df['Confirmed']<10].dropna()

```

```python
# 3. In which Region , maximum number of Confirmed cases were recorded?
df.groupby(['Confirmed','Region']).max()
# df['Confirmed'].min()
```

```python
#5. How many Confirmed , Deaths , Recovered cases were reported from India ?
dt=df[['Confirmed','Recovered','Deaths']]
dt[df['Region']=='India'].value_counts()
```

```python
# 6. Sort the entire Data wrt No of Confirmed Cases in Ascending Order 
df.sort_values([ 'Confirmed'],axis=0,ascending=[True],inplace=True)
df

```

```python
# 7. Sort the entire data wrt No of Recovered cases in Descending Order 
df.sort_values([ 'Recovered'],axis=0,ascending=[False],inplace=True)
df
```
